```python
import pandas as pd
import datetime


# Read in the original dataframe
df = pd.read_excel('macroeconomic_data/Macro_Economic_Indicators.xlsx')

# drop the last 5 rows
df = df.drop(df.tail(5).index)
df
```

```python
# Convert the 'year' column to datetime format
df['year'] = pd.to_datetime(df['year'], format='%Y')

# Set the 'year' column as the index
df.set_index('year', inplace=True)

# Resample the data by month and interpolate missing values
df_monthly = df.resample('M').interpolate()

# Reset the index to include the 'year' column as a regular column
df_monthly = df_monthly.reset_index()

# Rename the 'year' column to 'date'
df_monthly = df_monthly.rename(columns={'year': 'date'})

# Add a new column for the day, set to the first of the month
df_monthly['day'] = pd.to_timedelta(1, unit='D')

# Combine the 'date' and 'day' columns to create a new datetime column
df_monthly['datetime'] = df_monthly['date'] + df_monthly['day']

# Drop the 'date' and 'day' columns
df_monthly = df_monthly.drop(['date', 'day'], axis=1)
```
