# The Macro Economic Indicators are gathered from the Worldbank Database

https://datatopics.worldbank.org/world-development-indicators/

https://databank.worldbank.org/home.aspx

https://databank.worldbank.org/source/world-development-indicators

The World Development Indicators is a compilation of relevant, high-quality, and internationally comparable statistics about global development and the fight against poverty. The database contains 1,400 time series indicators for 217 economies and more than 40 country groups, with data for many indicators going back more than 50 years.

We've gathered data for you from 2005-2021 on:
- GDP
- Unemployment
- Land used for agriculture
- Renewable nergy consumption as share of the total energy consumption
- Fossil fuel consumption as share of the total energy consumption
- Energy use in kg of oil equivalents per capita


```python
import pandas as pd
import datetime

economic_indicators = pd.read_csv("macro_economic_indicators/macro_economic_indicators.csv", sep = ";", decimal=".")
economic_indicators["year"] = pd.to_datetime(economic_indicators["year"], format = '%Y').dt.date

for col in economic_indicators.columns:
    if col not in ['country_name', 'country_code', 'year', 'timecode']:        
        economic_indicators[col] = pd.to_numeric(economic_indicators[col])

economic_indicators.head(5)
```

```python
date_filter = datetime.date(year=2015,month=1,day=1)
country_code_filter = 'BEL'
```

# GDP

```python
economic_indicators[(economic_indicators['year']>=date_filter)&
                   (economic_indicators['country_code']==country_code_filter)]\
[['year', 'gdp_current_us', 'gdp_constant_2015_us']]\
.set_index('year').plot()
```

### Unemployment

```python
economic_indicators[(economic_indicators['year']>=date_filter)&
                   (economic_indicators['country_code']==country_code_filter)]\
[['year', 'unemployment_total']]\
.set_index('year').plot()
```

### Land usage

```python
economic_indicators[(economic_indicators['year']>=date_filter)&
                   (economic_indicators['country_code']==country_code_filter)]\
[['year', 'agricultural_land']]\
.set_index('year').plot()
```

### Energy Usage

```python
economic_indicators[(economic_indicators['year']>=date_filter)&
                   (economic_indicators['country_code']==country_code_filter)]\
[['year', 'renewable_energy_consumption_perc_of_total']]\
.set_index('year').plot()
```
